```python
#import the dataset
import statsmodels.api as sm
import pandas as pd
mtcars=sm.datasets.get_rdataset("mtcars","datasets",cache=True).data
df=pd.DataFrame(mtcars)
df.head()
```

Pearson correlation measures the linear relationship between X and Y variables

```python
import matplotlib.pyplot as plt
import seaborn as sns
corr=df.corr(method='pearson')
top_corr_features=corr.index
plt.figure(figsize=(10,10))
g=sns.heatmap(df[top_corr_features].corr(),annot=True,cmap="RdYlGn")
```

Kendall Correlation

```python
corr=df.corr(method='kendall')
top_corr_features=corr.index
plt.figure(figsize=(10,10))
g=sns.heatmap(df[top_corr_features].corr(),annot=True,cmap="Greens")
```

Spearman Correlation

```python
corr=df.corr(method='spearman')
top_corr_features=corr.index
plt.figure(figsize=(10,10))
g=sns.heatmap(df[top_corr_features].corr(),annot=True,cmap="Greens")
```

If we have lots of columns(features), we can only focus on the ones that are significantly statistically correlated, by setting a threshold(0.7)

```python
corr=df.corr()
kot=corr[(corr>=0.7) | (corr<=-0.7)]
plt.figure(figsize=(10,10))
g=sns.heatmap(kot,annot=True,cmap="RdYlGn")
```
